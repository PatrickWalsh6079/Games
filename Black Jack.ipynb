# Black Jack
This game is made for one human player and one computer player. It is a simple, text-based
version of Black Jack. There is a deck of 52 cards. Number cards are worth their numeric value,
jacks, kings, and queens are worth 10 points, and aces are worth 1 or 11. The player goes 
first. The player chooses either to 'hit' or 'stand'. Once the player is happy with their hand
and chooses 'stand', the turn goes to the computer dealer. The computer will then 'hit' until the 
computer has a sum of 17 or higher without exceeding 21, or the computer will lose by going over 
21 and busting. Player with the highest sum without going over 21 wins. If there is a tie, it is
called 'Push' and no money is exchanged that round.

The requirements are as follows:

The player can choose to 'stand' or 'hit'.
The deck will be represented using OOP.
The player gets to pick their betting amount.
The game must keep track of the player's total money.
The game must alert the player of a win or loss.
The game must incorporate Object Oriented Programming.


Object of the game:
Beat the dealer (computer player), either by getting 21 on their first two cards;
reach a final score higher than the dealer without exceeding 21; or
play until the dealer busts.
